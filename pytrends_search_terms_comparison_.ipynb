```python
import pandas as pd
from pytrends.request import TrendReq
pytrend = TrendReq(hl='en-US', tz=360)
import time
startTime = time.time()
```

```python
#colnames = ["keywords"]
#df = pd.read_csv("/content/Keywords.csv", names=colnames)
#df2 = df["keywords"].values.tolist()
#df2.remove("Keywords")
```

```python
df= pd.ExcelFile("/content/HumanRights_DAFramework_WorkingSheet_v1.1_31Ja2022.xlsx")
```

```python
data1= pd.read_excel(df, 'Master')
```

```python
data1
```

```python
data1.drop('Comments',axis=1)
```

```python
df1=data1.drop(['project_id','S.NO','Theme','Comments','Last Update'], axis=1)
```

```python
data2 = df1[(df1.Event == 'Discrimination') ]
data2
```

```python
colnames = ["Keyword(s)"]

df3 = data2["Keyword(s)"].values.tolist()
```

```python
df3
```

```python
dataset1 = []

for x in range(0,len(df3)):
     keywords = [df3[x]]
     pytrend.build_payload(
     kw_list=keywords,
     cat=0,
     timeframe='2020-01-31 2022-01-31',
     geo='')
     data = pytrend.interest_over_time()
     if not data.empty:
          data = data.drop(labels=['isPartial'],axis='columns')
          dataset1.append(data)
```

```python
result1= pd.concat(dataset1, axis=1)
result1.to_csv('search_trends_Discrimination.csv')
```
